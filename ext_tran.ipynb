```python
import pandas as pd
```

```python
#Load the data from a CSV file
```

```python
data = pd.read_csv("candidates.csv", sep=";")
data
```

A candidate is considered "Hired" if bouth values (code challenge score & technical interview Score) are >= 7
For that, im going to add a new column named "Hired" with true or false values (0 or 1), where 0 is gived when neighter code challenge score and technical interview Scores are <7 and 1 when bouth scores are >=7 


```python
#Here we define a function to determine if a candidate is hired based on their scores

def hired (row):
  if row["Code Challenge Score"] >= 7 and row["Technical Interview Score"] >= 7:
    return 1
  else:
    return 0
```

```python
#Apply the function to each row in the DataFrame and create a new column "Hired"
data["Hired"]= data.apply(hired, axis=1)
data
```

```python
data.info()
```

```python
data["Application Date"] = pd.to_datetime(data["Application Date"])
```

create tables for dimensions and facts to simplify the Load process

```python
#Im going to transform the data, and create the dimentions and facts tables

#DimCandidate 

dim_candidate = data[["First Name", "Last Name", "Email", "YOE"]].drop_duplicates().reset_index(drop=True)
dim_candidate["CandidateID"] = dim_candidate.index + 1

dim_candidate = dim_candidate[["CandidateID", "First Name", "Last Name", "Email", "YOE"]]
dim_candidate
```

```python
#DimCountry
dim_country = data[["Country"]].drop_duplicates().reset_index(drop=True)
dim_country["CountryID"] = dim_country.index + 1

dim_country = dim_country[["CountryID", "Country"]]
dim_country
```

```python
#DimDate 
dim_date = pd.DataFrame(data["Application Date"].drop_duplicates().reset_index(drop=True))
dim_date["DateID"] = dim_date.index + 1
dim_date["Application Date"] = pd.to_datetime(dim_date["Application Date"])
dim_date["Day"] = dim_date["Application Date"].dt.day
dim_date["Month"] = dim_date["Application Date"].dt.month
dim_date["Year"] = dim_date["Application Date"].dt.year
dim_date = dim_date.rename(columns={"Application Date": "ApplicationDate"})

dim_date = dim_date[["DateID", "ApplicationDate", "Day", "Month", "Year"]]
dim_date
```

```python
#DimSeniority
dim_seniority = pd.DataFrame(data["Seniority"].drop_duplicates().reset_index(drop=True))
dim_seniority["SeniorityID"] = dim_seniority.index + 1

dim_seniority = dim_seniority[["SeniorityID", "Seniority"]]
dim_seniority
```

```python
#DimTechnology
dim_tech = pd.DataFrame(data["Technology"].drop_duplicates().reset_index(drop=True))
dim_tech["TechnologyID"] = dim_tech.index + 1

dim_tech = dim_tech[["TechnologyID", "Technology"]]
dim_tech
```

```python
fact_applications = data.merge(dim_candidate, on=["First Name","Last Name","Email","YOE"]) \
                      .merge(dim_date, left_on="Application Date", right_on="ApplicationDate")  \
                      .merge(dim_country, on="Country") \
                      .merge(dim_seniority, on="Seniority") \
                      .merge(dim_tech, on="Technology")

fact_applications = fact_applications[[
    "CandidateID", "DateID", "SeniorityID", "TechnologyID","CountryID",
    "Code Challenge Score", "Technical Interview Score", "Hired"
]]
fact_applications["ApplicationID"] = fact_applications.index + 1

fact_applications = fact_applications[[
    "ApplicationID", "CandidateID", "DateID", "TechnologyID", "SeniorityID","CountryID",
    "Code Challenge Score","Technical Interview Score", "Hired"
]]

fact_applications
```

```python
# Guardar dimensiones y fact table como CSV para MySQL Workbench

dim_candidate.to_csv("dim_candidate.csv", index=False)
dim_date.to_csv("dim_date.csv", index=False)
dim_seniority.to_csv("dim_seniority.csv", index=False)
dim_tech.to_csv("dim_technology.csv", index=False)
dim_country.to_csv("dim_country.csv", index=False)
fact_applications.to_csv("fact_applications.csv", index=False)
```
